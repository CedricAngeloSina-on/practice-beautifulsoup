```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://salarygrade.ph/"
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, "html")
```

```python
table = soup.find("figure", class_ = "wp-block-table")
```

```python
table_titles = table.find_all("tr")[0]
df_titles = [title.text.strip() for title in table_titles]
```

```python
df_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=df_titles)
```

```python
df
```

```python
column_data = table.find_all("tr")[1:]
```

```python
for row in column_data:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
